```python
datagen = ImageDataGenerator(
    rotation_range=25,
    fill_mode='nearest',
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)
```

```python
image_2 = img_as_float(io.imread('./images/3.teacup-chihuahua-HL.jpg'))
datagen.fit(image_2)
```

```python
io.imshow(datagen.apply_transform(image_2,
{
    'theta':-25,
    'fill_mode':'nearest',
    'tx':0.1,
    'ty':0.1,
    'flip_horizontal':False
})) 
io.show()
```

```python
io.imshow(datagen.apply_transform(image_2,
{
    'theta':-25,
    'fill_mode':'nearest',
    'tx':0.1,
    'ty':0.1,
    'flip_horizontal':False
})) 
io.show()
```

```python
io.imshow(datagen.apply_transform(image_2,
{
    'theta':10,
    'fill_mode':'nearest',
    'tx':-0.1,
    'ty':0.05,
    'flip_horizontal':False
})) 
io.show()
```

```python
io.imshow(datagen.apply_transform(image_2,
{
    'theta':-10,
    'fill_mode':'nearest',
    'tx':-0.05,
    'ty':-0.07,
    'flip_horizontal':True
})) 
io.show()
```
